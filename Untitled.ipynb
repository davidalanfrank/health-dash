We require an extensive Redis store containing key-value pairs, where the keys encompass all observed key types linked to specific tables. The challenge involves categorizing a key extracted from the dexa data and determining the corresponding metric it is associated with. For instance, a key like "bmc_lean_mass_g" might be associated with a table named 'bmc_lean_mass.'

The subsequent task involves associating this data with a column in that particular table. This entails classifying a data point based on either its range or other available data points, determining the likelihood that it is related to a specific column. Once this association is established, we can proceed with the insertion of the data.
